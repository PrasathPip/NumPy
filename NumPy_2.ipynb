### Concept of axis in NumPy

### Sum - np.sum()

```python
np.sum(A) #returns the total sum of all elements in the array
```

```python

#when we only want the sum of elements along the columns or rows

# sum of elements along vertical direction - [0+3+6,1+4+7,2+5+8] = [9,12,15]
# sum of elements along horizontal direction - [0+1+2,3+4+5,6+7+8] = [3,12,21]
```

```python
A
```

```python
#for the above mentioned cases, the concept of axis is introduced in the sum method of numpy
'''np.sum(array,axis=0 or 1)'''
```

```python
'''--> Sum of elements in vertical direction :
VERTICAL DIRECTION - DIFFERENT ROWS ADDITION - AXIS 0


when the elements from many rows are added, that is sum along the vertical direction - axis = 0

0 belongs to row 0, 3 belongs to row 1, 6 belongs to row 2

[0+3+6,1+4+7,2+5+8] = [9,12,15]'''

np.sum(A,axis =0)
```

```python
'''--> Sum of elements in the horizontal direction :
HORIZONTAL DIRECTION - DIFFERENT COLUMN ADDITION - AXIS 1


when the elements from many columns are added, that is sum along the horizontal direction - axis = 1

0 belongs to column 0, 1 belongs to column 1, 2 belongs to column 2

[0+1+2,
3+4+5,
6+7+8] = [3,12,21]'''


np.sum(A,axis=1)
```

## Logical operators in numPy

### np.any - Logical OR

```python
A=np.array([False,False,True,False]) #atleast one TRUE
```

```python
"""np.any is similar to logical OR, It returns True,if "ANY" one value is True """
np.any(A)
```

```python
a=np.array([False,False,False,False]) #no TRUE value
```

```python
np.any(a)
```

```python
"""This holds true for integers also
in the case of integers, anything greater than 0 is treated as TRUE"""
B = np.array([0,0,3,0,0])
```

```python
np.any(B)
```

## np.all - Logical AND

```python
A=np.array([False,False,True,False]) #atleast one TRUE
```

```python
"""np.and is similar to logical AND, It returns True only if "ALL" the values are True """
np.all(A)
```

```python
a=np.array([True,True,True,True])
```

```python
np.all(a)
```

```python
#for integers
B=np.array([1,0,1,0]) #not all values are greater than 0 - Not all values are TRUE
```

```python
np.all(B)
```

```python
b=np.array([1,2,3,4]) #all values are greater than 0 - Hence all the values are TRUE
```

```python
np.all(b)
```

```python
A=np.array([1,2,3,4])
B=np.array([4,3,2,1])
np.any(A>B) #Returns TRUE since atleast one case satisfies the condition A>B, eg. 3>2 and 4>1
#np.any([F,F,T,T]) -->Returns True
```

```python
np.all(A>B) #Returns False since, not all the cases satisfy the condition A>B, eg. 1>4, 2>3
#np.all[F,F,T,T] -->Returns False
```

```python
A=np.array([4,5,6,7])
B=np.array([1,2,3,4])
np.all(A>B) #since ALL the cases satisfy the condition A>B, returns True
```

## Concept of Broadcasting

```python
"""When adding two arrays, of dissimilar dimensions, numpy tries it's level best to make the operations happen"""

A=np.array([[0,0,0],
  [10,10,10],
  [20,20,20],
  [30,30,30]])
```

```python
#CASE 1 --> ALL : Dimension of A and B are the same 4X3
B=np.array([[0,1,2],
  [0,1,2],
  [0,1,2],
  [0,1,2]])
```

```python
A+B #adds the corresponding element in each array
```

```python
#CASE 2 --> COLUMNS : The number of columns in B is equal to the number of columns in A
B=np.array([0,1,2])
```

```python
A+B #Still A+B produces the same result as, numpy tries to match [0,1,2] with the rows of A, temporarily creating
#  ([[0,1,2],
#   [0,1,2],
#   [0,1,2],
#   [0,1,2]])
```

```python
#CASE 3 
A=np.array([[0],
  [10],
  [20],
  [30]])
B=np.array([0,1,2])
```

```python
A+B #Still A+B produces the same result as, numpy tries to match [0,1,2] with the rows of A and to match [[0],[10],[20],[30]] with the columns of B
```

```python
A=np.array([[0,0,0],
  [10,10,10],
  [20,20,20],
  [30,30,30]])

B=np.array([[0,1],
  [0,1],
  [0,1],
  [0,1]])
```

```python
A+B  #Not compatible
# to be compatible, atleast on dimension should be 1
```
